# 3.2 Continuity and Differentiation

Limits are used to define continuity. A function is said to be continuous if it does not have any abrupt changes. A way to determine how a function changes is to use its derivative. The equation has the following form:

image.png

Where f' is the derivative of f.

Another insight given from the derivative is finding the extreme values of a function. If the derivative is zero for a certain value then depending on the change from increasing to decreasing or decreasing to increasing there will be a maximum or minimum point.

Let's look an at example of this:

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5,5, 100)
plt.plot(x, x**2)
plt.show()
```

The above plot shows the function F(x) = x^2. We can see there is a minimum point at x=0. But say we have a more complex function or we can't graph it. Then we can use the derivative of F(x) which would b F' = x to find the minimum point. The derivative is equal to zero at x=0 which is what the graph above also shows.

The same thing can be done for functions with more than one variable. In that case the function is set up such that each term involves the change of only one variable. Then using partial derivatives and the mean value theorem from above the properties of the function can be found.
