# 20.2 Finite Difference Approximating Derivatives
Using differences in the slope of a function the value around a certain point can be approximated. Three common ways to do so are the Forward, Backward, and Central difference formula.

The forward difference formula uses the line that connects the point of interest to the point to the right of it. It looks like this:

image.png

The backward difference formula uses the line that connects the point of interest to the point to the left of it. It looks like this:

image.png

The central difference formula uses the line that connects the points to the left and the right of the point of interest. It looks like this:

image.png

The forward and backward difference formulas are less accurate than the central difference one. This is because the central difference is symmetric while forward and backward are asymmetric and do not take into account how the function has and will change from past and future points.

Looking at an example function of Y = x^5 - 7x^2, we get the graph below. If we decrease the step size the accuracy will improve as well.

```python
import numpy as np
import matplotlib.pyplot as plt
h = 1
x = np.arange(0, 30, h)
y = x**5-7*x**2

forward_diff = np.diff(y)/h
x_diff = x[:-1:]
exact_solution = 5*(x_diff)**4-14*(x_diff)

plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Forward difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()
```

#20.4 Numerical Differentiation with Noise

The previous examples used a given function but sometimes, when using sensors for example, a function could have noise. Looking at Y = cos(2x). If the function had some noise say 0.1sin(50x), that would change how the function appears.

```python
import numpy as np
import matplotlib.pyplot as plt
h = 0.01
x = np.arange(0, 30, h)
y = np.cos(2*x)
noise = y + 0.1*np.sin(50*x)

plt.plot(x, noise, \
         label = 'Function with Noise')
plt.plot(x, y, \
         label = 'Exact Function')
plt.legend()
plt.show()
```
