# 21.4 Simpson's Rule

Simpson's Rule integrates over two consecutive subintervals by fitting a polynomial through each end point and the mid point. The equation looks like:


image.png

In order to use Simpson's rule, there must be an even number of subintervals. This method has an order of accuracy of 4 over the entire interval.

Looking at this example:

image.png

Let's find the approximate value using Simpson's Rule

```python
import numpy as np

a = 0
b = 6
n = 13
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x) + 3*x

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = (np.sin(6)+54) - I_simp

print(I_simp)
print(err_simp)
```
