#ODE Initial Value Problem Statement

An ODE is an ordinary differential equation. It is a relationship between the derivative and independent variable in one dimension.

There are different solutions to an ODE. One is a particular solution, it satisfies the differential equation and some number of known values of the solution. Very commonly the known values of the solution are the initial values. Determining the exact solution algebraically for a lot of ODEs is quite difficult. That is where different approzimation methods come in handy.


#Reduction of Order

Before using the different approximation methods, many differential equations must have their order reduced. The approximation methods are for first order differential equations so anything higher must be put into first order form in order to be estimated.

This is done by taking derivatives with respect to one variable and placing them in matrix form.

#The Euler Method
Starting from a known value explicit euler formula allows us to estimate the next value and eventually the entire function. It follows the form:

image.png

Let's look at an example using this method. Assuming f' = cos(x) and the initial condition is f(0) = 1, then the exact solution should be f = 1 + sin(x).

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Define parameters
f = lambda t, s: np.cos(t) # ODE
h = 0.1 # Step size
t = np.arange(0, 5 + h, h) # Numerical grid
s0 = 1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, 1+np.sin(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlim([0,5])
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

That was the explicit euler method. The implicit euler method uses a linear approximation around the next point in order to estimate it, i.e it is not written in terms of tj but instead tj+1:
