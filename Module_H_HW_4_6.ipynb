#22.6 Python ODE Solvers

While we went over the manual formulas for integrating to solve a differential equation python also has built in ODE solvers. They use the ODE, interval of integration, initial state, and a given method to solve the ODE.

Let's look at an example. Assuming dx/dt = sin(2t) and initial condition is -0.5, the solution should be Y = -0.5cos(2t)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp


F = lambda t, s: np.sin(2*t)

t_eval = np.arange(0, np.pi, 0.1)
sol = solve_ivp(F, [0, np.pi], [-0.5], t_eval=t_eval)

plt.figure(figsize = (12, 4))
plt.subplot(121)
plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('S(t)')
plt.subplot(122)
plt.plot(sol.t, sol.y[0] + 0.5*np.cos(2*sol.t))
plt.xlabel('t')
plt.ylabel('S(t) - sin(t)')
plt.tight_layout()
plt.show()

```
