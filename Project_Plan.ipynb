#Introduction to the Problem
Accurately collecting data is a critical challenge in engineering systems, particularly for temperature tracking. Temperature monitoring plays a vital role in applications ranging from medical devices such as thermometers to large-scale climate agreements like the Paris Agreement.

While highly accurate measurement systems exist, they are often prohibitively expensive. Engineers must balance manufacturing costs with equipment accuracy when designing systems. In situations where continuous data collection is limited, estimating missing values becomes essential. For instance, if a measurement tool only records data every 30 seconds, estimating the temperature at the 15-second mark can improve data accuracy without additional hardware costs. These estimations must be as precise as possible to ensure the system functions as intended. For this research paper, I will use Lagrange Interpolation and Spline Interpolation to estimate missing temperature values and compare the results with Polynomial Regression for accuracy assessment.
#Related Work
Various methods for temperature estimation have been explored in previous research. One common approach involves single-channel algorithms that correlate satellite data with temperature readings. However, these methods can be unreliable due to atmospheric variations that affect accuracy.

Another effective technique is Lagrange Interpolation combined with ground thermometers, which has shown promising results. Notably, one study demonstrated that Lagrange Interpolation outperformed several industry-standard algorithms in numerical and engineering design optimization, underscoring its potential for improved accuracy. This highlights the need for further exploration and comparison with other interpolation methods such as Spline Interpolation and Polynomial Regression.

#Experiment Setups
The dataset used in this study consists of global monthly temperature averages from 1850 to 2022. For the experiment, data from the years 2000 to 2022 will be utilized while deliberately excluding the year 2015. The dataset will be cleaned and normalized, and missing values will be handled accordingly to ensure consistency across methods.

The three methods (Lagrange Interpolation, Spline Interpolation, and Polynomial Regression) will be applied to the available temperature data to construct predictive models. The models will then be used to estimate the temperature values for the excluded year 2015. The estimated values for 2015 will be compared against the actual recorded values using absolute error.

#Expected Results
Based on the characteristics of the three models, Spline Interpolation is expected to yield the most accurate temperature estimations due to its ability to fit smooth transitions between data points.

Lagrange Interpolation, which fits a single polynomial over the entire dataset, may not accurately capture temperature variations and could introduce large errors, especially for larger datasets.

Polynomial Regression is likely to perform worse than Spline Interpolation due to the risk of overfitting and high oscillations between data points, which may not accurately reflect real-world temperature trends.
